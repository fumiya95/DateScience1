```python
import pandas as pd
import matplotlib.pyplot as plt

# データの準備
data = {
    '日付': ["2024/10/10", "2024/10/11", "2024/10/12", "2024/10/13", "2024/10/14", "2024/10/15"],
    '売上': [242952, 181293, 105584, 195055, 137457, 206921],
    '来客数': [309, 226, 120, 238, 163, 211]
}

# データフレームの作成
df = pd.DataFrame(data)

# 日付列を日付型に変換
df['日付'] = pd.to_datetime(df['日付'])

# 日付列をインデックスに設定
df.set_index('日付', inplace=True)

# 売上データと来客数データを同一のグラフにプロットするための準備
fig, ax1 = plt.subplots()

# 売上データを赤色の円マーカーでプロット
ax1.plot(df.index, df['売上'], marker='o', color='red', label='売上')
ax1.set_xlabel('日付')  # X軸に「日付」を設定
ax1.set_ylabel('売上', color='red')  # Y軸に「売上」を設定し、ラベルの色を赤に
ax1.tick_params(axis='x', rotation=45)  # X軸のティックラベルを45度回転

# 来客数データを別のY軸を使用してプロット
ax2 = ax1.twinx()  # 共共有軸を持つ別のY軸の補佐を作成
ax2.plot(df.index, df['来客数'], marker='s', color='skyblue', label='来客数')
ax2.set_ylabel('来客数', color='skyblue')  # 補佐に「来客数」を設定し、ラベルの色をスカイブルーに
ax2.set_ylim(0, df['来客数'].max() * 2)  # 来客数のY軸の範囲を合わせるために最大値の2倍まで設定

# グラフのタイトルとレイアウトの調整
plt.title('売上と来客数の日付ごとの推移')  # グラフのタイトルを設定
fig.tight_layout()  # サブプロット間のスペースを自動調整

# 売上と来客数の凡例をまとめて表示
lines1, labels1 = ax1.get_legend_handles_labels()  # ax1の凡例のハンドル(グラフのマーカーやラインのスタイルを示すオブジェクト)とラベル(凡例のテキスト)を取得
lines2, labels2 = ax2.get_legend_handles_labels()  # ax2の凡例のハンドルとラベルを取得
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right')  # ax1とax2の凡例を結合して、グラフの右上に表示

# グラフの表示
plt.xticks(rotation=45)  # X軸のティックラベルを45度回転して表示
plt.show()  # グラフを表示
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# データの準備
data = {
    '日付': ["2024/10/10", "2024/10/11", "2024/10/12", "2024/10/13", "2024/10/14", "2024/10/15"],
    '売上': [242952, 181293, 105584, 195055, 137457, 206921],
    '来客数': [309, 226, 120, 238, 163, 211]
}

# データフレームの作成
df = pd.DataFrame(data)

# 日付列を日付型に変換
df['日付'] = pd.to_datetime(df['日付'])

# 2つのサブプロットを含む図を作成します。図のサイズは幅10インチ、高さ8インチです。
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# 上のサブプロット(ax1)に売上データをプロットします。
# データポイントは赤色の円マーカー('o')で表示されます。
ax1.plot(df['日付'], df['売上'], label='売上', marker='o', color='red')
ax1.set_xlabel('日付')  # X軸には「日付」を表示
ax1.set_ylabel('売上', color='red')  # Y軸には「売上」を表示し、ラベルの色を赤に設定
ax1.set_title('売上の日付ごとの推移')  # サブプロットのタイトルを設定
ax1.tick_params(axis='y', labelcolor='red')  # Y軸のティックラベルの色を赤に設定
ax1.legend(loc='upper left')  # 凡例を左上に表示
ax1.tick_params(axis='x', rotation=45)

# 下のサブプロット(ax2)に来客数データをプロットします。
# データポイントは緑色の正方形マーカー('s')で表示されます。
ax2.plot(df['日付'], df['来客数'], label='来客数', marker='s', color='green')
ax2.set_xlabel('日付')  # X軸には「日付」を表示
ax2.set_ylabel('来客数', color='green')  # Y軸には「来客数」を表示し、ラベルの色を緑に設定
ax2.set_ylim(0, df['来客数'].max() * 2)  # Y軸の範囲を0から来客数の最大値の2倍まで設定
ax2.set_title('来客数の日付ごとの推移')  # サブプロットのタイトルを設定
ax2.tick_params(axis='y', labelcolor='green')  # Y軸のティックラベルの色を緑に設定
ax2.legend(loc='upper left')  # 凡例を左上に表示
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()  # サブプロット間のスペースを自動調整

# グラフを表示します。
plt.show()
```
