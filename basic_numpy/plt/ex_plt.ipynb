```python
import matplotlib.pyplot as plt
import matplotlib.font_manager
import pandas as pd
plt.rcParams['font.family'] = 'Hiragino Sans'
```

```python
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value': [23, 17, 35, 29, 12],
    'Value2': [35, 23, 45, 21, 33]
}

```

```python
plt.bar(data['Category'], data['Value'], color='skyblue', edgecolor='black')
plt.title('Bar Graph')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
df = pd.DataFrame(data)
plt.plot(df['Category'], df['Value2'], marker='o', color='green', label='Value2')
plt.title('Line Graph')
plt.xlabel('Category')
plt.ylabel('Value2')
plt.legend()
plt.show()
```

```python
# 散布図の作成
plt.scatter(data['Value'], data['Value2'], color='red')

# タイトルとラベルの設定
plt.title('Scatter Plot')
plt.xlabel('Value')
plt.ylabel('Value2')

# グラフの表示
plt.show()
```

```python
# カテゴリを数値インデックスに変換
index = range(len(data['Category']))

# 散布図の作成
plt.scatter(index, data['Value'], color='red', label='Value')
plt.scatter(index, data['Value2'], color='blue', label='Value2')

# タイトルとラベルの設定
plt.title('Color coding')
plt.xlabel('Index')
plt.ylabel('Value')

# 凡例の追加
plt.legend()

# グラフの表示
plt.show()
```
