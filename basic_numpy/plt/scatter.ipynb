```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager
```

```python
plt.rcParams['font.family'] = 'Hiragino Sans'
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

```python
plt.scatter(x, y) 
```

```python
# グラフのカスタマイズ
plt.title('3つの値から散布図の作成')
plt.xlabel('X軸のラベル')
plt.ylabel('Y軸のラベル')

# グラフの表示
plt.show()
```

```python
data = pd.read_csv('temperature_ice_cream_sales.csv')
data
```

```python
plt.scatter(data['気温'], data['アイスの売れた数'], s=5, c="red", alpha=0.5, marker='^')
plt.xlabel('気温')

plt.ylabel('アイスの売れた数')

plt.title('Temperature vs Ice Cream Sales')
plt.show()
```

```python
plt.scatter(data['気温'], data['アイスの売れた数'],c="red",label='今年')
plt.scatter(data['気温']-2, data['アイスの売れた数']-30, c="blue", label='去年')
plt.xlabel('気温')
plt.ylabel('アイスの売れた数')
plt.title('Temperature vs Ice Cream Sales')
plt.legend()
plt.show()
```

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris_japanese = iris.rename(columns={

    'sepal_length': 'がく片長',

    'sepal_width': 'がく片幅',

    'petal_length': '花弁長',

    'petal_width': '花弁幅',

    'species': '種'

})
```

```python
sns.pairplot(iris_japanese, hue='種')

plt.suptitle('Irisデータセットの散布図行列', y=1.02)

plt.show()
```
