_**問題16**_<br>
df1とdf2を縦に結合し、出力してください。

＜実行結果＞<br>
　列1　	列2<br>
0　	A	　10<br>
1　	B	　20<br>
2　	C　	30<br>
3　	a　	1000<br>
4	　b	　2000<br>
5	　c　	3000


```python
data1 = [
    ['A', 10],
    ['B', 20],
    ['C', 30]
]

data2 = [
    ['a', 1000],
    ['b', 2000],
    ['c', 3000]
]

# データフレームの作成
df1 = pd.DataFrame(data1, columns=['列1', '列2'])
df2 = pd.DataFrame(data2, columns=['列1', '列2'])
concatenated_df = pd.concat([df1, df2], ignore_index=True)

print(concatenated_df)
```

_**問題20**_<br>
sample_data.csvファイルを、以下の画像のように棒グラフにして表示してください。<br>
＜条件＞<br>
1. x軸に「Category」、y軸に「Value」をセットしてください
2. タイトルを「Sample Bar Graph」としてください。
3. y軸のタイトルを「Value」にしてください。
4. x軸のタイトルを「Category」にしてください
5. 棒グラフの色を「skyblue」にしてください

<img src='output.png' width="50%">

```python


# CSVファイルを読み込む
df = pd.read_csv('sample_data.csv')

# 棒グラフを作成
plt.bar(df['Category'], df['Value'], color='skyblue')

# グラフのタイトルとラベルを設定
plt.title('Sample Bar Graph')
plt.xlabel('Category')
plt.ylabel('Value')

# グラフを表示
plt.show()
```
