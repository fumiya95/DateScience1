```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt

import matplotlib.font_manager

plt.rcParams['font.family'] = 'Hiragino Sans' 
```

```python
stores = ['A店', 'B店', 'C店', 'D店', 'E店']

proceeds = [100, 200, 150, 250, 300]



# DataFrameの作成

df = pd.DataFrame({

    '店舗': stores,

    '売上': proceeds

})



df
```

```python
plt.bar(df['店舗'], df['売上'])



# グラフのタイトルとラベルの設定

plt.title('Basicな棒グラフ')

plt.xlabel('店舗')

plt.ylabel('売上')



# グラフの表示

plt.show()
```

```python
# 水平棒グラフの作成

plt.barh(df['店舗'], df['売上'])



# グラフのタイトルとラベルの設定

plt.title('Basic水平棒グラフ')

plt.xlabel('売上')

plt.ylabel('店舗')



# グラフの表示

plt.show()
```

```python
plt.bar(df['店舗'], df['売上'], color='skyblue', edgecolor='blue')



# グラフのタイトルとラベルの設定

plt.title('Custom!!')

plt.xlabel('店舗')

plt.ylabel('売上')



# グラフの表示

plt.show()
```

```python
stores = ['A店', 'B店', 'C店', 'D店', 'E店']

proceeds = [100, 200, 150, 250, 300]

costs = [50, 80, 60, 110, 180]



# DataFrameの作成

df = pd.DataFrame({

    '店舗': stores,

    '売上': proceeds,

    'コスト':costs

})



df
```

```python
import numpy as np



# 棒グラフの幅を設定

bar_width = 0.35



# 売上データを青色の棒グラフで表示

plt.bar(df.index, df['売上'], bar_width, label='売上', color='blue')



# コストデータを赤色の棒グラフで表示（売上データの隣に表示）

plt.bar(df.index + bar_width, df['コスト'], bar_width, label='コスト', color='red')



# グラフのタイトルを設定

plt.title('複数棒グラフ')



# x軸のラベルを設定

plt.xlabel('店舗')



# y軸のラベルを設定

plt.ylabel('売上')



# x軸の目盛りを設定し、x軸に店舗名を指定（棒グラフの中心に店舗名が来るように調整）

plt.xticks(df.index + bar_width / 2, df['店舗'])



# 凡例を表示

plt.legend()



# グラフを表示

plt.show()
```
